# Absenteeism - Visualisation and Analysis

```python
# import the relevant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
# We can override the default matplotlib styles with those of Seaborn
sns.set()## Importing the relevant libraries
```

```python
# Load the data from a .csv file
data_processed = pd.read_csv(os.path.join(os.path.pardir, 'src', 'features', 'predicted_absenteeism.csv'))
df = data_processed.copy()
df
```

#### Age vs Probability

```python
plt.scatter(df['Age'], df['Probability'].apply(lambda x: x*100))
plt.xlabel('Age',size=18)
plt.ylabel('Probability in %',size=18)
# plt.xlim(6,13)
# plt.ylim(6,13)
plt.show()
```

#### Reasons vs Probability

```python
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, sharey=True, figsize =(15,9))

ax1.scatter(df['Reason_1'], df['Probability'].apply(lambda x: x*100), s=10, c='b', marker="s")
ax1.set_title('Probability vs Reason 1')

ax2.scatter(df['Reason_2'], df['Probability'].apply(lambda x: x*100), s=10, c='r', marker="s")
ax2.set_title('Probability vs Reason 2')

ax3.scatter(df['Reason_3'], df['Probability'].apply(lambda x: x*100), s=10, c='y', marker="s")
ax3.set_title('Probability vs Reason 3')

ax4.scatter(df['Reason_4'], df['Probability'].apply(lambda x: x*100), s=10, c='g', marker="s")
ax4.set_title('Probability vs Reason 4')

plt.show()
```

#### Transportation Expense, and Children vs Probability

```python
plt.figure(figsize=(6,6))
plt.scatter(df['Transportation Expense'],
            df['Probability'].apply(lambda x: x*100),
            c=df['Children'], cmap='rainbow', s=df['Children']*100,)
plt.xlabel('Transportation Expense',size=18)
plt.ylabel('Probability in %',size=18)

plt.show()

# The larger the circle the more the children
```
